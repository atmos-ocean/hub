```python
import numpy as np
import matplotlib.pyplot as plt

NAME='u'; TITLE='u=u1+u2+u3+u4'
#NAME='u1'; TITLE='u1'
#NAME='u2'; TITLE='u2'
#NAME='u3'; TITLE='u3'
#NAME='u4'; TITLE='u4'

t=2

LCLR="black"
LSTY="solid"

A1=0.3; c1=1.0; L1=7; k1=2.0*np.pi/L1
LCLR1=LCLR #"r"
LSTY1=LSTY #"dashed"

A2=0.3; c2=1.1; L2=5.5; k2=2.0*np.pi/L2
LCLR2=LCLR #"r"
LSTY2=LSTY #"solid"

A3=0.3; c3=1.2; L3=3.8; k3=1.0*np.pi/L3
LCLR3=LCLR #"r"
LSTY3=LSTY #"solid"

A4=0.3; c4=0.; L4=1.7; k4=1.0*np.pi/L4
LCLR4=LCLR #"r"
LSTY4=LSTY #"solid"

x = np.linspace(-12, 12, 300)

plt, ax=plt.subplots(figsize=(6, 3), dpi=300, tight_layout=True)

ax.set_xlim(-10, 10)
ax.set_ylim(-1.3, 1.3)

u1=A1*np.cos(k1*(x-c1*t))
u2=A2*np.cos(k2*(x-c2*t))
u3=A3*np.cos(k3*(x-c3*t))
u4=A4*np.cos(k4*(x-c4*t))

u=u1+u2+u3+u4

ax.plot(x,u, color=LCLR, ls=LSTY, label="t="+str(t))

#ax.plot(x,u1, color=LCLR1, ls=LSTY1, label="t="+str(t))
#ax.plot(x,u2, color=LCLR2, ls=LSTY2, label="t="+str(t))
#ax.plot(x,u3, color=LCLR3, ls=LSTY3, label="t="+str(t))
#ax.plot(x,u4, color=LCLR4, ls=LSTY4, label="t="+str(t))

start, end = ax.get_xlim()
ax.xaxis.set_ticks(np.arange(start, end, 2))

ax.axhline(0, ls='-', c='b', lw=0.5)
ax.axvline(0, ls='-', c='b', lw=0.5)

ax.tick_params(labelsize=16)

#ax.legend(fontsize=14)
ax.legend(frameon=False)
ax.set_xlabel('x', fontsize=16)
ax.set_ylabel(NAME, fontsize=16)
ax.set_title(TITLE,fontsize=16)

plt.show()

# https://qiita.com/eriho/items/21302ab7e95a2721b9f8
# matplotlibにはplt.plotをベースしたmatlab準拠の方法と、
# ax.plotをベースとしたオブジェクト指向の方法がある
#
# 最初からax.plot方式に慣れておいた方がいい
# 
```
