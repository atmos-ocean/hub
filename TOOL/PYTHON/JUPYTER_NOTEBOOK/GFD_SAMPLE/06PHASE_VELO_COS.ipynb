```python
import numpy as np
import matplotlib.pyplot as plt

TITLE="Progressive Wave"

t1=0
t2=2

LCLR1="r"
LSTY1="dashed"

LCLR2="r"
LSTY2="solid"

c=1.0
#xc=-8
L=12
k=2.0*np.pi/L

x = np.linspace(-12, 12, 300)


plt, ax=plt.subplots(figsize=(6, 3), dpi=300, tight_layout=True)
#plt.figure(figsize=(6, 3), dpi=300)

ax.set_xlim(-10, 10)
ax.set_ylim(-1.3, 1.3)



ax.plot(x, np.cos(k*(x-c*t1)), color=LCLR1, ls=LSTY1, label="t="+str(t1))
ax.plot(x, np.cos(k*(x-c*t2)), color=LCLR2, ls=LSTY2, label="t="+str(t2))



start, end = ax.get_xlim()
ax.xaxis.set_ticks(np.arange(start, end, 2))

ax.axhline(0, ls='-', c='b', lw=0.5)
ax.axvline(0, ls='-', c='b', lw=0.5)


ax.tick_params(labelsize=16)


ax.legend(fontsize=14)
#ax.legend(frameon=False)
ax.set_xlabel('x', fontsize=16)
ax.set_ylabel('u', fontsize=16)
ax.set_title(TITLE,fontsize=16)

plt.show()

# https://qiita.com/eriho/items/21302ab7e95a2721b9f8
# matplotlibplt.plotmatlab
# ax.plot
#
# ax.plot
# 
```
