```python
import numpy as np
import matplotlib.pyplot as plt

k = np.linspace(0.1, 1, 100)
#x = np.linspace(-2*np.pi, 2*np.pi)

plt.figure(figsize=(6, 4), dpi=300)

#plt.xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi, ],["-2π","-π", "0", "π","2π",])

o=-1/k

plt.plot(k, o, color='r', ls='-', label="")

plt.xlim(0, 1)
#plt.ylim(-1.3, 1.3)

plt.axhline(0, ls='-', c='b', lw=0.5)
plt.axvline(0, ls='-', c='b', lw=0.5)

plt.legend(fontsize=16)
plt.tick_params(labelsize=16)



#plt.legend()
plt.xlabel(r'$k$', fontsize=16)
plt.ylabel(r'$\omega$', fontsize=16)
plt.title('Rossby wave dispersion relation')

plt.show()
```
