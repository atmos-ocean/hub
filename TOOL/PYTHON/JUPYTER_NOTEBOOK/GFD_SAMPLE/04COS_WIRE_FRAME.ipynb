```python
from mpl_toolkits.mplot3d import Axes3D   
import matplotlib.pyplot as plt 
import numpy as np
import matplotlib.ticker as ticker

#fig = plt.figure() #プロット領域の作成

fig = plt.figure(figsize=(10,10), dpi=300)
ax = fig.add_subplot(111, projection='3d')
#ax = fig.gca(projection='3d') #プロット中の軸の取得。gca は"Get Current Axes" の略。

x = np.arange(-2, 2, 0.025) # x点として[-2, 2]まで0.05刻みでサンプル
t = np.arange(0, 4, 0.025)  # y点として[-2, 2]まで0.05刻みでサンプル
x, t = np.meshgrid(x, t)  # 上述のサンプリング点(x,y)を使ったメッシュ生成

k=2*np.pi/1
omega=2*np.pi/2

u = np.cos((k*x - omega*t))
ax.set_zlim(-1.0,1.0)
ax.set_xlabel('x',fontsize=22)
ax.set_ylabel('t',fontsize=22)
ax.set_zlabel('u',fontsize=22)
plt.tick_params(labelsize=16)

ax.view_init(elev=30, azim=290)
# elevはz軸方向から見た仰角，azimはx, y軸方向の方位角を指定．
# matplotlib mplot3d API

tick_spacing=1
tick_spacingz=0.5

ax.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))
ax.yaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))
ax.zaxis.set_major_locator(ticker.MultipleLocator(tick_spacingz))

surf=ax.plot_surface(x, t, u, rstride=1, cstride=1, cmap='coolwarm', linewidth=0.3) 
# 曲面のプロット。rstrideとcstrideはステップサイズ，cmapは彩色，linewidthは曲面のメッシュの線の太さ，をそれぞれ表す。

cbar = fig.colorbar(surf, shrink = 0.6)

cbar.set_label("u", fontsize=20)


plt.show() # 絵の出力。
```
