```python
import numpy as np
import matplotlib.pyplot as plt

TITLE='(Rightward)'
LCLR='r'
t=15.0
c=1.0
xc=-8
L=5
k=2.0*np.pi/L

x = np.linspace(-10, 10, 300)
#x = np.linspace(-2*np.pi, 2*np.pi)

plt.figure(figsize=(6, 2), dpi=300)

#plt.xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi, ],["-2π","-π", "0", "π","2π",])

plt.plot(x, 1.0/(np.cosh(k*(x-xc)-c*t)**2), color=LCLR, ls='-', label="")
#plt.plot(x, 1.0/(np.cosh(k*((x-c*t))**2, color='r', ls='-', label="t="+str(t))

plt.xlim(-10, 10)
#plt.xlim(-2*np.pi, 2*np.pi)
plt.ylim(-2.3, 2.3)

plt.axhline(0, ls='-', c='b', lw=0.5)
plt.axvline(0, ls='-', c='b', lw=0.5)

plt.legend(fontsize=16)
plt.tick_params(labelsize=16)


plt.legend(frameon=False)
plt.xlabel('x', fontsize=16)
plt.ylabel('u', fontsize=16)
plt.title('t='+str(t)+" "+TITLE,fontsize=20)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

TITLE='(Superposition)'
LCLR1='r'
LSTY1="dashed"
LCLR2='purple'
LSTY2="dashed"
LCLR3='black'
LSTY3="solid"


t=15.0

c1=1.0
xc1=-8

c2=-1.0
xc2=8

L=5
k=2.0*np.pi/L

x = np.linspace(-10, 10, 300)
#x = np.linspace(-2*np.pi, 2*np.pi)

plt.figure(figsize=(6, 2), dpi=300)

#plt.xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi, ],["-2π","-π", "0", "π","2π",])

plt.plot(x, 1.0/(np.cosh(k*(x-xc1)-c1*t)**2)+1.0/(np.cosh(k*(x-xc2)-c2*t)**2), color=LCLR3, linestyle=LSTY3, label="")

plt.plot(x, 1.0/(np.cosh(k*(x-xc1)-c1*t)**2), color=LCLR1, linestyle=LSTY1, label="")

plt.plot(x, 1.0/(np.cosh(k*(x-xc2)-c2*t)**2), color=LCLR2, linestyle=LSTY2, label="")


plt.xlim(-10, 10)
#plt.xlim(-2*np.pi, 2*np.pi)
plt.ylim(-2.3, 2.3)

plt.axhline(0, ls='-', c='b', lw=0.5)
plt.axvline(0, ls='-', c='b', lw=0.5)

plt.legend(fontsize=16)
plt.tick_params(labelsize=16)


plt.legend(frameon=False)
plt.xlabel('x', fontsize=16)
plt.ylabel('u', fontsize=16)
plt.title('t='+str(t)+" "+TITLE,fontsize=20)

plt.show()
```
