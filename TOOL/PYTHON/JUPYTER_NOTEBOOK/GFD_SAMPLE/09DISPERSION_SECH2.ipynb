```python
import numpy as np
import matplotlib.pyplot as plt

NAME='u1+u2'; TITLE='u1+u2'
#NAME='u1'; TITLE='u1'

LCLR="black"
LSTY="solid"

A1=0.5; T1=2; L1=12.; o1=2.0*np.pi/T1; k1=2.0*np.pi/L1; c1=o1/k1
LCLR1="r"
LSTY1="dashed"

A2=0.3; T2=1; L2=12.; o2=2.0*np.pi/T2; k2=2.0*np.pi/L2; c2=o2/k2
LCLR2="b"
LSTY2="dashed"

print(c1); print(c2)

xmin=-5; xmax=30; xint=5
x = np.linspace(xmin, xmax, 300)

plt, ax=plt.subplots(figsize=(6, 3), dpi=300, tight_layout=True)

ax.set_xlim(xmin, xmax)
ax.set_ylim(-2, 2)

t=2

u1=A1*1.0/(np.cosh(k1*x-o1*t))*2
u2=A2*1.0/(np.cosh(k2*x-o2*t))*2
u=u1+u2

ax.plot(x,u, color=LCLR, ls=LSTY, label="")

#ax.plot(x,u1, color=LCLR1, ls=LSTY1, label="t="+str(t))
#ax.plot(x,u2, color=LCLR2, ls=LSTY2, label="t="+str(t))

start, end = ax.get_xlim()
ax.xaxis.set_ticks(np.arange(start, end, xint))

ax.axhline(0, ls='-', c='b', lw=0.5)
ax.axvline(0, ls='-', c='b', lw=0.5)

ax.tick_params(labelsize=16)

ax.legend(fontsize=0,loc=3)
ax.legend(frameon=False)
ax.set_xlabel('x', fontsize=18)
ax.set_ylabel(NAME, fontsize=18)
ax.set_title("t="+str(t)+" "+TITLE ,fontsize=25)

plt.show()

# https://qiita.com/eriho/items/21302ab7e95a2721b9f8
# matplotlibにはplt.plotをベースしたmatlab準拠の方法と、
# ax.plotをベースとしたオブジェクト指向の方法がある
#
# 最初からax.plot方式に慣れておいた方がいい
# 
```
