```python
import numpy as np
import matplotlib.pyplot as plt

y = np.linspace(0, 100, 101)
#x = np.linspace(-2*np.pi, 2*np.pi)

plt.figure(figsize=(6, 4), dpi=300)

#plt.xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi, ],["-2π","-π", "0", "π","2π",])

T=24-0.1*y


plt.plot(y, T, color='r', ls='-', label="")
plt.grid()

plt.xlim(0, 100)
#plt.ylim(-1.3, 1.3)

#plt.axhline(0, ls='-', c='b', lw=0.5)
#plt.axvline(0, ls='-', c='b', lw=0.5)

plt.legend(fontsize=16)
plt.tick_params(labelsize=16)



#plt.legend()
plt.xlabel(r'$y [km]$', fontsize=16)
plt.ylabel(r'$T [C]$', fontsize=16)
plt.title('Meridional distribution')

plt.show()
```
